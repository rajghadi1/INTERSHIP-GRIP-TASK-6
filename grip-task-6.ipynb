# Modeling

```python
x = df.iloc[:, 0:-1].values # Independent variables
y = df.iloc[:, -1].values  # Dependent or Target variable (Iris-Species)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Decision Tree Classification Algorithm

**Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, but mostly it is preferred for solving Classification problems. It is a tree-structured classifier, where internal nodes represent the features of a dataset, branches represent the decision rules and each leaf node represents the outcome.**

```python

classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
model=classifier.fit(X_train, y_train)
model
```

# **Prediction**

```python
print(classifier.predict(sc.transform([[7.0,3.2,4.7,1.4]])))
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

# Accuracy

```python
cm = confusion_matrix(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred)*100))
```

```python
pd.DataFrame(np.c_[X_test,y_test,y_pred],columns=['SepalLengthCm',	'SepalWidthCm',	'PetalLengthCm',	'PetalWidthCm',	'Species','predication'])
```

# K-Means 

```python
x=df.iloc[:,0:4].values
```

```python
from sklearn.cluster import AgglomerativeClustering

groups = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')

labels=groups .fit_predict(x)
labels
```

```python

# Visualising the clusters
plt.figure(figsize=(10,10))
plt.scatter(x[labels==0,0],x[labels==0,1],s=100,c='red',label='Iris-setosa')
plt.scatter(x[labels==1,0],x[labels==1,1],s=100,c='blue',label='Iris-versicolour')
plt.scatter(x[labels==2,0],x[labels==2,1],s=100,c='green',label='Iris-virginica')

# Plotting the centroids of the clusters
#plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='Centroids')
plt.legend()
```

```python
df['Species'] = pd.Categorical(df["Species"])
df["Species"] = df["Species"].cat.codes
df["Species"]
```

```python
from sklearn.metrics import classification_report

target_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

print(classification_report(df['Species'],groups.labels_,target_names=target_names))
```
